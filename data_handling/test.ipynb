```python
import pandas as pd

df = pd.read_csv('cancer.csv')

# drop last column
df.drop(df.columns[-1], axis=1, inplace=True)
df.dropna(inplace=True)

# rename column diagnosis to malignant
df.rename(columns={'diagnosis': 'malignant'}, inplace=True)

# convert malignant to 0 and 1
df['malignant'] = df['malignant'].map({'M': 1, 'B': 0})

# drop id column
df.drop('id', axis=1, inplace=True)

# take 500 rows of data randomly
df = df.sample(n=500, random_state=1)

# count malignant and benign
print(df['malignant'].value_counts())

df.reset_index(drop=True, inplace=True)

# save to data.csv
df.to_csv('data.csv', index=False)
```
